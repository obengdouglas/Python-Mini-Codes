#3. Write a sorting function without using the list.sort function (descending order)


```python
lst=[24, 55, 78, 64, 25, 12, 22, 11, 1, 2, 44, 3, 122, 23, 34]
n_list = []

while lst:
  minimum=lst[0]
  for x in lst:
    if x>minimum:
      minimum=x
  n_list.append(minimum)
  lst.remove(minimum)

print(n_list)
```

#4. Write a Python program to print Fibonacci Series

```python
def f(n):
  if n == 0: return 0
  elif n == 1: return 1
  else: return f(n-1)+f(n-2)

for n in range(0,15):
  print(f(n))
```

#6. Write a Python program to check whether a string is a Palindrome or not (Hint: A palindrome is a word, phrase, number or sequence of words that reads the same backward as forward. e.g. Madam)

```python
def Palindrome(s):
  rev = ''.join(reversed(s))

  if (s==rev):
    return True
  return False

Palindrome('madam')
```

#7. Write a Python program to print set of duplicates in a list

```python
l = [1,2,3,4,4,5,5,6,1]
n = []
for i in l:
  if l.count(i)>1:
    #and i not in n:
    n.append(i)
print(set(n)) #set returns unique elements

```

#9. Given an array arr[] of n elements, write a Python function to search a given element x in arr[].

```python
def search(arr,x):
  for i in range(len(arr)):
    if arr[i]==x:
      return i
  print('Not in list')

l = [1,2,3,4,4,5,5,6,1]

search(l,6)

```
