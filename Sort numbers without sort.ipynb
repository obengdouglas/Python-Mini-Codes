Write a sorting function without using the list.sort function (descending order)


```python
lst=[24, 55, 78, 64, 25, 12, 22, 11, 1, 2, 44, 3, 122, 23, 34]
n_list = []

while lst:
  minimum=lst[0]
  for x in lst:
    if x>minimum:
      minimum=x
  n_list.append(minimum)
  lst.remove(minimum)

print(n_list)
```
