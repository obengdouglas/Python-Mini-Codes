```python
def vowel_count(str):
    count=0
    vowel = set(["a", "e", "i", "o", "u", "A", "E", "O", "I", "U"])
    for alphabet in str:
        if alphabet in vowel:
            count = count+1
    print("No of vowels: ", count)
```
